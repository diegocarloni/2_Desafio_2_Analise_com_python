# 3. Qual foi o pais que teve queda de "Generosity" ao longo dos anos ? Classificar o TOP 5, nos últimos 10 anos

## Tratamento/Exclusão de valores nulos do campo Generosidade

```python
questaotres = original[original.Generosity.notnull()].copy()
```

## Após tramento, Checar campos "nulos" em Generosidade

```python
questaotres.isnull().sum()
```

# Criar um dataframe com os países do ano 2010

```python
gen2010 = questaotres[questaotres['year'] == 2010][['Country name', 'Generosity']]
gen2010
```

# Criar um dataframe com os países do ano 2020

```python
gen2020 = questaotres[questaotres['year'] == 2020][['Country name', 'Generosity']]
gen2020
```

# Criar um dataframe com Generosity dos anos 2010 e 2020

```python
delta_gen = pd.merge(gen2010, gen2020, how = 'inner', on = 'Country name')
delta_gen
```

 # Renomear as colunas

```python
delta_gen = delta_gen.rename(columns={'Generosity_x': 'Gen_2010', 'Generosity_y': 'Gen_2020'})
delta_gen
```

# Transformação em lista - dados 2010

```python
Gen_2010 = delta_gen['Gen_2010'].tolist()
Gen_2010
```

# Transformação em lista - dados 2020

```python
Gen_2020 = delta_gen['Gen_2020'].tolist()
Gen_2020
```

# For - Filtrar valores em "queda"

```python
a = 0
resultado = []
for i in range(0,len(delta_gen)):
    if Gen_2010[a] > Gen_2020[a]:
        rang = Gen_2010[a] - Gen_2020[a]
        resultado.append(rang)
    else:
        resultado.append(0)
    a +=1
```

# Adicionar a coluna "Range"

```python
delta_gen['Range']=resultado
```

```python
delta_gen
```

# Buscar os 5 países com maior diferença entre os índices

```python
delta_gen.nlargest(5,'Range')
```

# Gráfico de barras aninhadas

```python
plt.figure(figsize=(10,5))

# Quantidade de vendas para o Produto A
Turkey = [0.623, 0.510]

# Quantidade de vendas para o Produto B
Iran = [0.651, 0.600]

# Cria eixo x para produto A e produto B com uma separação de 0.25 entre as barras
x1 =  np.arange(len(Turkey))
x2 = [x + 0.25 for x in x1]

# Plota as barras
plt.bar(x1, Turkey, width=0.25, label = 'Turkey', color = 'b')
plt.bar(x2, Iran, width=0.25, label = 'Iran', color = 'y')

# coloca o nome dos meses como label do eixo x
anos = ['2005','2020']
plt.xticks([x + 0.25 for x in range(len(Turkey))], anos)

# inseri uma legenda no gráfico
plt.legend()

plt.title("Análise Liberdade entre o primeiro e último registro")
plt.show()
```
